# Operações com os dados de todos os anos

#### Juntei os dados relacionados à licitação, licitantes, propostas e itens já tratados nos outros notebooks de todos os anos com o comando: cat 2016/arquivo_2016.txt 2017/arquivo_2017.txt 2018/arquivo_2018.txt 2019/arquivo_2019.txt > Todos_os_Anos.txt

# Removendo a coluna Unnamed, que são os índices dos outros arqruivos

```python
licitacoes = licitacoes.drop(licitacoes.columns[0], axis=1)
```

```python
licitacoes
```

# Checando se há linhas duplicadas

```python
print("Duplicated values: ", licitacoes.duplicated().sum())
```

```python
duplicateRowsDF = licitacoes[licitacoes.duplicated()]
duplicateRowsDF
```

# Ao concatenar o cabeçalho estava duplicado, aqui vou removê-los deixando apenas o primeiro

```python
licitacoes.drop_duplicates(keep=False,inplace=True) 
licitacoes
```

# Convertendo as colunas de valores em float

```python
licitacoes['VL_LICITACAO'] = pd.to_numeric(licitacoes['VL_LICITACAO'],errors='coerce')
licitacoes['VL_TOTAL_PROPOSTA'] = pd.to_numeric(licitacoes['VL_TOTAL_PROPOSTA'],errors='coerce')
licitacoes['VL_TOTAL_HOMOLOGADO'] = pd.to_numeric(licitacoes['VL_TOTAL_HOMOLOGADO'],errors='coerce')
licitacoes['VL_UNITARIO_HOMOLOGADO'] = pd.to_numeric(licitacoes['VL_UNITARIO_HOMOLOGADO'],errors='coerce')
licitacoes.dtypes
```

# Checando a estatística dos valores numéricos

```python
licitacoes.describe()
```

# One Hot Encoding

### Para transformar os objetos em valores numéricos

```python
le = preprocessing.LabelEncoder()
for column in licitacoes.columns:
    if(licitacoes[column].dtypes=='object'):
        licitacoes[column] = le.fit_transform(licitacoes[column])
        
licitacoes.head()
```

# Removendo as colunas que possuem valores iguais em todas as instâncias

```python
licitacoes = licitacoes.drop(['TP_OBJETO', 'TP_RESULTADO_PROPOSTA'], axis=1)
```

# Removendo os outliers através do z-score

```python
from scipy import stats
df_sem_outliers = licitacoes[(np.abs(stats.zscore(licitacoes)) < 3).all(axis=1)]
df_sem_outliers
```

# Checando a correlação entre os atributos

```python
corr = df_sem_outliers.corr()
corr.style.background_gradient(cmap = 'coolwarm')
```

### Removendo as colunas que possuem valores iguais em todas as instâncias

```python
df_sem_outliers = df_sem_outliers.drop(['TP_DOCUMENTO', 'TP_LICITACAO', 'CD_TIPO_FASE_ATUAL'], axis=1)
```

```python
corr = df_sem_outliers.corr()
corr.style.background_gradient(cmap = 'coolwarm')
```

# Criando modelo de regressão para prever o VL_LICITACAO

```python
# import all libraries and dependencies for machine learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing

import warnings
warnings.filterwarnings('ignore')
```

# Separando os dados em treinamento e teste

```python
# Train test split
X = df_sem_outliers.drop(['VL_LICITACAO'], axis = 1)
y = df_sem_outliers['VL_LICITACAO']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

# Testando a regressão linear

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_train, y_train)

y_pred = reg.predict(X_test)

# Print out the mean absolute error (mae)
print("r2 score : ",r2_score(y_test,y_pred))
```

# Testando a regressão por KNN

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=3)
neigh.fit(X_train, y_train)

y_pred = neigh.predict(X_test)

# Print out the mean absolute error (mae)
print("r2 score : ",r2_score(y_test,y_pred))
```

# Testando a regressão por Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 200, max_depth= 5)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
print("r2 score : ",r2_score(y_test,y_pred))
```

# Como o Random Forest deu melhor que os outros, é possível que o XGBoost dê melhor que todos, visto que ele é um ensemble de árvores e decisões, então vou testá-lo

```python
params = {'n_estimators': 100,
          'max_depth': 4,
          'min_samples_split': 5,
          'learning_rate': 0.05,
          'loss': 'ls',
          'random_state': 0}
```

```python
gbr = GradientBoostingRegressor(**params)
gbr.fit(X_train,y_train)
y_pred = gbr.predict(X_test)

print("r2 score : ",r2_score(y_test,y_pred))
```
